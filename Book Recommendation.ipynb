```python
# Import Libraries
import numpy as np
import pandas as pd
```

```python
# Load Dataset
df = pd.read_csv("books.csv")

df.head()
```

```python
# Select Important Columns
df = df[['book_id','title','author','genre','rating']]
df = df.copy()

df.head()
```

```python
# Handle Missing Values
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
# Create Tags (Feature Engineering)
df['tags'] = df['title'] + " " + df['author'] + " " + df['genre']

df.head()
```

```python
# Convert to Lowercase
df['tags'] = df['tags'].apply(lambda x: x.lower())

df.head()
```

```python
# Convert to Lowercase
df['tags'] = df['tags'].apply(lambda x: x.lower())

df.head()
```

```python
# Final DataFrame
new_df = df[['book_id','title','tags']]

new_df.head()
```

```python
# Vectorization
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()

vectors.shape
```

```python
# Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)

similarity.shape
```

```python
# Recommendation Function
def recommend(book):
    book = book.lower()

    if book not in new_df['title'].str.lower().values:
        print("Book not found!")
        return

    index = new_df[new_df['title'].str.lower() == book].index[0]

    distances = list(enumerate(similarity[index]))
    books_list = sorted(distances, key=lambda x: x[1], reverse=True)[1:6]

    print("Top 5 Recommendations:\n")
    for i, b in enumerate(books_list):
        print(i+1, ".", new_df.iloc[b[0]].title)
```

```python
# Test
recommend("Harry Potter")
```

```python
# Save Model
import pickle

pickle.dump(new_df, open('books.pkl','wb'))
pickle.dump(similarity, open('books_similarity.pkl','wb'))
```
